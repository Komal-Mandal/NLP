```python
import string
exclude = string.punctuation
print(exclude)
```

```python
def rem_punc1(text):
    if isinstance(text, str):
        return text.translate(str.maketrans("", "", exclude))
    else:
        return text  
```

```python
df["tweet"].apply(rem_punc1)
```
