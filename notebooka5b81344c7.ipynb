```python
import re

def remove_html_tags(input_text):
    """
    Removes HTML tags from a given string.
    
    Parameters:
        input_text (str): The string containing HTML tags.
    
    Returns:
        str: The string with HTML tags removed.
    """
    # Regular expression to match HTML tags
    tag_re = re.compile(r"<[^>]*>")
    
    # Replace HTML tags with an empty string
    cleaned_text = tag_re.sub("", input_text)
    
    return cleaned_text

# Example usage
html_text = "<p>This is <strong>bold</strong> and <em>italic</em> text.</p>"
print(remove_html_tags(html_text))

```

```python
df["review"] = df["review"].apply(remove_html_tags)
```

```python
import re

def remove_urls(input_text):
    """
    Removes URLs from a given string.
    
    Parameters:
        input_text (str): The string containing URLs.
    
    Returns:
        str: The string with URLs removed.
    """
    # Regular expression to match URLs
    url_re = re.compile(r"https?://[^\s]+|www\.[^\s]+")
    
    # Replace URLs with an empty string
    cleaned_text = url_re.sub("", input_text)
    
    return cleaned_text

# Example usage
text_with_urls = "Check out this website: https://example.com and also visit www.test.com for more info."
print(remove_urls(text_with_urls))
```
